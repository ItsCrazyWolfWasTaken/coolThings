```python
from keras.datasets import mnist
import numpy as np
import random
(train_X, train_y), (test_X, test_y) = mnist.load_data()

train_X = train_X.astype(np.float32) / 255.0
test_X = test_X.astype(np.float32) / 255.0
```

```python
def relu(num):
    return max(0, num)
def original_weights(num_nodes):
    weights = []
    for i in range(len(num_nodes)):
        weights.append([])
        for j in range(num_nodes[i]):
            weights[i].append(random.random())
    return weights
def forward_prop(input_data, weights):
    output = []
    temp = []
    output.append([])
    for i in range(len(weights[0])):
        output[0].append(input_data[i] * weights[i])
    for j in range(len(weights)-1):
        output.append([])
        for k in range(len(weights[j+1])):
            output[j+1].append()
    return output
```
